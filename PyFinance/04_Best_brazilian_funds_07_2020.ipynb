# Fundos com melhores rendimentos de julho de 2020



```
import pandas as pd
pd.set_option("display.max_colwidth", 150)
```

```
def busca_informes_cvm(ano, mes):
  url = 'http://dados.cvm.gov.br/dados/FI/DOC/INF_DIARIO/DADOS/inf_diario_fi_{:02d}{:02d}.csv'.format(ano,mes)
  return pd.read_csv(url, sep=';')
```

```
def busca_cadastro_cvm(ano, mes, dia):
  url = 'http://dados.cvm.gov.br/dados/FI/CAD/DADOS/inf_cadastral_fi_{}{:02d}{:02d}.csv'.format(ano, mes, dia)
  return pd.read_csv(url, sep=';', encoding='ISO-8859-1') #Terceiro parâmetro passado é para correcão na leitura do CSV por conta da acentuação BRL
```

```
informes_diarios = busca_informes_cvm(2020,7)
```

```
informes_diarios
```

```
cadastro_cvm = busca_cadastro_cvm(2020,7,1)
```

```
cadastro_cvm
```

Dropa a linha da base de dados se o valor não estiver na outra base. 

Nesse caso tinham vários fundos de 'informes_diarios' quem não tinham CNPJ na base 'cadastro_cvm'

```
informes_diarios_confirmados = informes_diarios[informes_diarios['CNPJ_FUNDO'].isin(cadastro_cvm['CNPJ_FUNDO'])]
informes_diarios_confirmados
```

```
minimo_cotistas = 100
fundos = informes_diarios_confirmados[informes_diarios_confirmados['NR_COTST'] >= minimo_cotistas].pivot(index='DT_COMPTC', columns='CNPJ_FUNDO', values=['VL_TOTAL',	'VL_QUOTA',	'VL_PATRIM_LIQ',	'CAPTC_DIA',	'RESG_DIA'])
```

```
fundos
```

```
cotas_normalizadas = fundos['VL_QUOTA'] / fundos['VL_QUOTA'].iloc[0]
```

```
cotas_normalizadas
```

```
melhores = pd.DataFrame()
melhores['retorno(%)'] = (cotas_normalizadas.iloc[-1].sort_values(ascending=False)[:50] - 1) * 100
melhores
```

```
for cnpj in melhores.index:
  fundo = cadastro_cvm[cadastro_cvm['CNPJ_FUNDO'] == cnpj]
  melhores.at[cnpj, 'Fundo de Investimento'] = fundo['DENOM_SOCIAL'].values[0]
  melhores.at[cnpj, 'Classe'] = fundo['CLASSE'].values[0]
  melhores.at[cnpj, 'PL'] = fundo['VL_PATRIM_LIQ'].values[0]
```
